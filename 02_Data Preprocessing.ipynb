```python
# Importing necessary libraries for encoding categorical variables
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
# one-hot encoding for age_description it removes the bias of the age_description column
ohe = OneHotEncoder()
encoded = ohe.fit_transform(df[['age_description']])
# encoded.toarray()
```

```python
# to get the feature names after one-hot encoding
ohe.get_feature_names_out()
```

```python
# Create a new DataFrame with the one-hot encoded features
df_new = pd.DataFrame(
encoded.toarray(),columns=ohe.get_feature_names_out()
)
```
